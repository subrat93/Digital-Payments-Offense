```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc, confusion_matrix
from sklearn.metrics import average_precision_score
from sklearn.metrics import roc_curve
from matplotlib import pyplot
from sklearn import metrics
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import pickle


from IPython.display import display

def formatted_print(metric_name, header=False, metric_value=None, linebreak=False, header_type='h3'):
    html_str = ""
    if header:
        html_str += '<hr><' + header_type + '>'
    html_str += '<b>' + metric_name 
    if not header:
        html_str += ': </b>'
    else:
        html_str += '</b></' + header_type + '>'
    if linebreak:
        html_str += '<br />'
    if metric_value:
        html_str += str(metric_value)
    if header:
        html_str += '<hr>'
    display({'text/html': html_str}, raw=True)
```

```python
def check_train_metrics(model, X_train, y_train):
    formatted_print("Training Metrics", header=True, header_type='h1', linebreak=True)
    train_predicted = model.predict(X_train)
    train_predicted_proba = model.predict_proba(X_train)
    
    # Calculate the metrics for default threshold
    formatted_print("With Threshold = 0.5", header=True, header_type='h2')
    train_accuracy = metrics.accuracy_score(y_train, train_predicted)
    formatted_print("Accuracy Score (Train)", metric_value = train_accuracy)    
    train_f1 = f1_score(y_train, train_predicted)
    formatted_print("F1 Score (Train)", metric_value = train_f1)
    # calculate precision-recall AUC
    precision, recall, thresholds = precision_recall_curve(y_train, train_predicted_proba[:,1])
    mdl_auc = auc(recall, precision)
    formatted_print("Area Under Curve (Train)", metric_value = mdl_auc)
    train_precision = metrics.precision_score(y_train, train_predicted)
    train_recall = metrics.recall_score(y_train, train_predicted)
    formatted_print("Precision (Train)", metric_value = train_precision)
    formatted_print("Recall (Train)", metric_value = train_recall)
    formatted_print("Classification Report (Train)")
    print(metrics.classification_report(y_train, train_predicted))
    
    # Calculate the best threshold
    tn, fp, fn, tp = confusion_matrix(y_train, train_predicted).ravel()
#     best_train_threshold = get_metric_and_best_threshold_from_roc_curve(y_train, train_predicted_proba[:,1])
    best_train_acc, best_train_threshold = get_metric_and_best_threshold_from_pr_curve(precision, recall, thresholds, tp+fn, fp+tn)
    formatted_print("Best Threshold (Train)", metric_value = best_train_threshold)
    
    # Calculate the metrics for best threshold
    formatted_print("With Threshold = {}".format(best_train_threshold), header=True, header_type='h2')
    train_predicted_ts = (train_predicted_proba[:,1] >= best_train_threshold).astype(int)
    train_accuracy_ts = metrics.accuracy_score(y_train, train_predicted_ts)
    formatted_print("Accuracy Score (Train)", metric_value = train_accuracy_ts)    
    train_f1_ts = f1_score(y_train, train_predicted_ts)
    formatted_print("F1 Score (Train)", metric_value = train_f1_ts)
    # calculate precision-recall AUC
    precision, recall, thresholds = precision_recall_curve(y_train, train_predicted_proba[:,1])
#     mdl_auc_ts = auc(recall, precision)
#     formatted_print("Area Under Curve (Train)", metric_value = mdl_auc_ts)
    train_precision_ts = metrics.precision_score(y_train, train_predicted_ts)
    train_recall_ts = metrics.recall_score(y_train, train_predicted_ts)
    formatted_print("Precision (Train)", metric_value = train_precision_ts)
    formatted_print("Recall (Train)", metric_value = train_recall_ts)
#     formatted_print("Classification Report (Train)")
#     print(metrics.classification_report(y_train, train_predicted_ts))

    return train_accuracy, train_f1, mdl_auc, train_precision, train_recall, train_accuracy_ts, train_f1_ts,\
    train_precision_ts, train_recall_ts, best_train_threshold
    
    
```

```python
def cross_validation(model, X_train, y_train):
#     formatted_print("Cross-validation", header=True)
    formatted_print("Cross-validation", header=True, header_type='h1', linebreak=True)
    # Create k-Fold cross-validation
    kf = KFold(n_splits=5, shuffle=True, random_state=123)
    # Do k-fold cross-validation
    cv_results = cross_val_score(model, # Model
                                 X_train, # Feature matrix
                                 y_train, # Target vector
                                 cv=kf, # Cross-validation technique
                                 scoring="accuracy", # Loss function
                                 n_jobs=-1) # Use all CPU scores
    # print("Mean Accuracy in CV Training: ", cv_results.mean())
    formatted_print("Mean Accuracy in CV Training", metric_value = cv_results.mean())
    predicted = cross_val_predict(model, X_train, y_train, cv=10)
    cv_acc = metrics.accuracy_score(y_train, predicted)
    formatted_print("Accuracy Score (CV)", metric_value = cv_acc)
    cv_f1 = f1_score(y_train, predicted)
    formatted_print("F1 Score (CV)", metric_value = cv_f1)
    # calculate precision-recall AUC
    cv_precision = metrics.precision_score(y_train, predicted)
    cv_recall = metrics.recall_score(y_train, predicted)
    formatted_print("Precision (CV)", metric_value = cv_precision)
    formatted_print("Recall (CV)", metric_value = cv_recall)
    formatted_print("Classification Report (CV)", linebreak=True)
    print(metrics.classification_report(y_train, predicted))
    return cv_acc, cv_f1, cv_precision, cv_recall
```

```python
def check_test_metrics(model, X_test, y_test, best_train_threshold):
#     formatted_print("Test Metrics", header=True)
    formatted_print("Test Metrics", header=True, header_type='h1', linebreak=True)
    # predict probabilities
    probs = model.predict_proba(X_test)
    # keep probabilities for the positive outcome only
    probs = probs[:, 1]
    # predict class values
    yhat = model.predict(X_test)
    
    # Calculate the metrics for default threshold
    formatted_print("With Threshold = 0.5", header=True, header_type='h2')
    test_acc = metrics.accuracy_score(y_test, yhat)
    formatted_print("Accuracy Score(Test)", metric_value=test_acc)
    test_f1 = f1_score(y_test, yhat)
    formatted_print("F1 Score (Test)", metric_value = test_f1)
    # calculate precision-recall AUC
    test_precision = metrics.precision_score(y_test, yhat)
    test_recall = metrics.recall_score(y_test, yhat)
    formatted_print("Precision (Test)", metric_value = test_precision)
    formatted_print("Recall (Test)", metric_value = test_recall)
    formatted_print("Classification Report (Test)", linebreak=True)
    print(metrics.classification_report(y_test, yhat))
    # calculate precision-recall curve
    precision, recall, thresholds = precision_recall_curve(y_test, probs)
    # calculate F1 score
    f1 = f1_score(y_test, yhat)
    # calculate precision-recall AUC
    lg_auc = auc(recall, precision)
    # calculate average precision score
    ap = average_precision_score(y_test, probs)
#     formatted_print("Other Test Metrics", linebreak=True)
#     print(' f1_score=%.3f \n auc=%.3f \n average_precision=%.3f' % (f1, lg_auc, ap))
    # calculate roc curve
    fpr, tpr, thresholds = roc_curve(y_test, probs)
    pyplot.figure(figsize=(5,4))
    # plot no skill
    pyplot.plot([0, 1], [0, 1], linestyle='--')
    # plot the roc curve for the model
    pyplot.plot(fpr, tpr, marker='.')
    pyplot.title('AUC-ROC Curve')
    # show the plot
    pyplot.show()

    pyplot.figure(figsize=(5,4))
    # plot no skill
    pyplot.plot([0, 1], [0.5, 0.5], linestyle='--')
    # plot the precision-recall curve for the model
    pyplot.plot(recall, precision, marker='.')
    pyplot.title('Precision-Recall Curve')
    # show the plot
    pyplot.show()
    
    formatted_print("Confusion Matrix (Default Threshold)", linebreak=True)
    show_confusion_matrix(y_test, yhat)
    
    # Calculate the metrics for best threshold
    formatted_print("With Threshold = {}".format(best_train_threshold), header=True, header_type='h2')
    test_predicted_ts = (probs >= best_train_threshold).astype(int)
    test_accuracy_ts = metrics.accuracy_score(y_test, test_predicted_ts)
    formatted_print("Accuracy Score (Test)", metric_value = test_accuracy_ts)    
    test_f1_ts = f1_score(y_test, test_predicted_ts)
    formatted_print("F1 Score (Test)", metric_value = test_f1_ts)
    # calculate precision-recall AUC
    precision, recall, thresholds = precision_recall_curve(y_test, probs)
    mdl_auc_ts = auc(recall, precision)
    formatted_print("Area Under Curve (Test)", metric_value = mdl_auc_ts)
    test_precision_ts = metrics.precision_score(y_test, test_predicted_ts)
    test_recall_ts = metrics.recall_score(y_test, test_predicted_ts)
    formatted_print("Precision (Test)", metric_value = test_precision_ts)
    formatted_print("Recall (Test)", metric_value = test_recall_ts)
#     formatted_print("Classification Report (Train)")
#     print(metrics.classification_report(y_train, train_predicted_ts))
    formatted_print("Confusion Matrix (Best Threshold)", linebreak=True)
    show_confusion_matrix(y_test, test_predicted_ts)

    return test_acc, test_f1, lg_auc, test_precision, test_recall, test_accuracy_ts, test_f1_ts,\
    test_precision_ts, test_recall_ts
```

```python
def tune_hyperparameter(model, param_dict, X_train, y_train, ht_scoring):
#     formatted_print("Hyperparameter Tuning", header=True)
    formatted_print("Hyperparameter Tuning", header=True, header_type='h1', linebreak=True)
    gridF = RandomizedSearchCV(model, param_dict, cv = 3, verbose = 1, scoring = ht_scoring, random_state = 28,
                          n_jobs = -1)
    bestF = gridF.fit(X_train, y_train)
    formatted_print("Best Hyperparameter Tuning Score", metric_value=bestF.best_score_)
    formatted_print("Best Hyperparameter Tuning params", metric_value=bestF.best_params_)
    return bestF.best_estimator_
```

```python
def auto_run(model_name, model, X_train, y_train, X_test, y_test):
    
    # Cross-validate using training data
    cv_acc, cv_f1, cv_precision, cv_recall = cross_validation(model, X_train, y_train)
    
    # Get Training metrics
    train_accuracy, train_f1, train_auc, train_precision, train_recall, \
    train_accuracy_best_threshold, train_f1_best_threshold, train_precision_best_threshold, train_recall_best_threshold, \
    best_train_threshold = check_train_metrics(model, X_train, y_train)
    
    # Get Test metrics
    test_accuracy, test_f1, test_auc, test_precision, test_recall, test_accuracy_best_threshold, test_f1_best_threshold, \
    test_precision_best_threshold, test_recall_best_threshold = check_test_metrics(model, X_test, y_test, best_train_threshold)
    
    # Get the concordance metrics for test data
    concordance, discordance, tied = calculate_concordance_ratio(model, X_test, y_test)
    
```

```python
from sklearn.metrics import confusion_matrix
def show_confusion_matrix(y_test, y_pred):
    
    C = confusion_matrix(y_test,y_pred)
    class_labels = ['No Offense', 'Offense']
    assert C.shape == (2,2), "Confusion matrix should be from binary classification only."
    
    # true negative, false positive, etc...
    tn = C[0,0]; fp = C[0,1]; fn = C[1,0]; tp = C[1,1];

    NP = fn+tp # Num positive examples
    NN = tn+fp # Num negative examples
    N  = NP+NN

    fig = plt.figure(figsize=(8,8))
    ax  = fig.add_subplot(111)
    ax.imshow(C, interpolation='nearest')

    # Draw the grid boxes
    ax.set_xlim(-0.5,2.5)
    ax.set_ylim(2.5,-0.5)
    ax.plot([-0.5,2.5],[0.5,0.5], '-k', lw=2)
    ax.plot([-0.5,2.5],[1.5,1.5], '-k', lw=2)
    ax.plot([0.5,0.5],[-0.5,2.5], '-k', lw=2)
    ax.plot([1.5,1.5],[-0.5,2.5], '-k', lw=2)

    # Set xlabels
    ax.set_xlabel('Predicted Label', fontsize=16)
    ax.set_xticks([0,1,2])
    ax.set_xticklabels(class_labels + [''])
    ax.xaxis.set_label_position('top')
    ax.xaxis.tick_top()
    # These coordinate might require some tinkering. Ditto for y, below.
    ax.xaxis.set_label_coords(0.34,1.06)

    # Set ylabels
    ax.set_ylabel('True Label', fontsize=16, rotation=90)
    ax.set_yticklabels(class_labels + [''],rotation=90)
    ax.set_yticks([0,1,2])
    ax.yaxis.set_label_coords(-0.09,0.65)


    # Fill in initial metrics: tp, tn, etc...
    ax.text(0,0,
            'True Neg: %d\n(Num Neg: %d)'%(tn,NN),
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))

    ax.text(0,1,
            'False Neg: %d'%fn,
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))

    ax.text(1,0,
            'False Pos: %d'%fp,
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))


    ax.text(1,1,
            'True Pos: %d\n(Num Pos: %d)'%(tp,NP),
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))

    # Fill in secondary metrics: accuracy, true pos rate, etc...
    ax.text(2,0,
            'False Pos Rate: %.2f'%(fp / (fp+tn+0.)),
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))

    ax.text(2,1,
            'True Pos Rate: %.2f'%(tp / (tp+fn+0.)),
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))

    ax.text(2,2,
            'Accuracy: %.2f'%((tp+tn+0.)/N),
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))

    ax.text(0,2,
            'Neg Pre Val: %.2f'%(1-fn/(fn+tn+0.)),
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))

    ax.text(1,2,
            'Pos Pred Val: %.2f'%(tp/(tp+fp+0.)),
            va='center',
            ha='center',
            bbox=dict(fc='w',boxstyle='round,pad=1'))


    plt.tight_layout()
    plt.show()
```

```python
def get_metric_and_best_threshold_from_pr_curve(
    precision, recall, thresholds, num_pos_class, num_neg_class
):
    tp = recall * num_pos_class
    fp = (tp / precision) - tp
    tn = num_neg_class - fp
    acc = (tp + tn) / (num_pos_class + num_neg_class)

    best_threshold = thresholds[np.argmax(acc)]
    return np.amax(acc), best_threshold
```

```python
def get_metric_and_best_threshold_from_roc_curve(y_test, y_pred):
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)
    J = tpr - fpr
    ix = np.argmax(J)
    best_thresh = thresholds[ix]
    return best_thresh
```

```python
def calculate_concordance_ratio(model, X_test, y_test):
    Probability = model.predict_proba(X_test)
    Probability1 = pd.DataFrame(Probability)
    Probability1.columns = ['Prob_0','Prob_1']
    TruthTable = pd.merge(y_test, Probability1[['Prob_1']], how='inner', left_index=True, right_index=True)
    zeros = TruthTable[(TruthTable['MULTIPLE_OFFENSE']==0)].reset_index().drop(['index'], axis = 1)
    ones = TruthTable[(TruthTable['MULTIPLE_OFFENSE']==1)].reset_index().drop(['index'], axis = 1)

    from bisect import bisect_left, bisect_right

    zeros_list = sorted([zeros.iloc[j,1] for j in zeros.index])
    zeros_length = len(zeros_list)
    disc = 0
    ties = 0
    conc = 0
    for i in ones.index:
        cur_conc = bisect_left(zeros_list, ones.iloc[i,1])
        cur_ties = bisect_right(zeros_list, ones.iloc[i,1]) - cur_conc
        conc += cur_conc
        ties += cur_ties
    pairs_tested = zeros_length * len(ones.index)
    disc = pairs_tested - conc - ties

#     print("Pairs = ", pairs_tested)
#     print("Conc = ", conc)
#     print("Disc = ", disc)
#     print("Tied = ", ties)
    concordance = conc/pairs_tested
    discordance = disc/pairs_tested
    ties_perc = ties/pairs_tested
    
    formatted_print("Concordance %", metric_value = concordance)
    formatted_print("Discordance %", metric_value = discordance)
    formatted_print("Tied %", metric_value = ties_perc)
    
    return concordance, discordance, ties_perc
```
